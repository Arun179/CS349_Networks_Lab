```python
import numpy as np
fig, ax = plt.subplots(figsize=(10,10))
w = 0.25

r1 = np.arange(5)
r2 = [x + w for x in r1]
r3 = [x + w for x in r2]
plt.bar(r1,data['avg. RTT 1'],width=w)
ax.bar(r2,data['avg. RTT 2'],width=w)
ax.bar(r3,data['avg. RTT3'],width=w)

plt.xticks([r + w for r in range(5)], data['Hostname'],fontsize=15)
```
